```python
def find_album(num):
    hdr = { 'User-Agent' : 'Mozilla/5.0 (Windows NT 6.1; Win64; x64)' }
    url = "https://music.bugs.co.kr/track/"
    url = url + str(num)
    try:
        html = urlopen(urllib.request.Request(url, headers=hdr))
    except urllib.error.HTTPError as e:
        print(e)
        print("=============error %d ============="%num)
    else:
        html = BeautifulSoup(html.read(), "html.parser")
        title = html.find('header', {'class': 'sectionPadding pgTitle noneLNB'}).get_text()
        title = del_title(title)
        lyrics = html.find('div', {'class': 'lyricsContainer'}).get_text()
        
        writer.writerow([title, lyrics])
```
