```python
def FindPetition(num):
    url = "https://www1.president.go.kr/petitions/"
    url = url + str(num)

    html = urlopen(url)
    bsObj = BeautifulSoup(html.read(), "html.parser")

    progress = bsObj.find("div",{"class":"petitionsView_progress"}).get_text()
    title = bsObj.find("h3",{"class":"petitionsView_title"}).get_text()
    count = bsObj.find("h2",{"class":"petitionsView_count"}).get_text()
    petition_overview = bsObj.find("div",{"class":"View_write"}).get_text()
    days = bsObj.find("ul",{"class":"petitionsView_info_list"}).get_text()
    days = days.split("\n")
    category = days[1]
    sdays = days[2]
    edays = days[3]
    person = days[4]

    return [category, sdays, edays, person, progress, title, count, petition_overview]
```

```python
import re

def FindPetition(num):
    url = "https://www1.president.go.kr/petitions/"
    url = url + str(num)

    html = urlopen(url)
    bsObj = BeautifulSoup(html.read(), "html.parser")

    progress = bsObj.find("div",{"class":"petitionsView_progress"}).get_text()
    title = bsObj.find("h3",{"class":"petitionsView_title"}).get_text()
    count = bsObj.find("h2",{"class":"petitionsView_count"}).get_text()
    petition_overview = bsObj.find("div",{"class":"View_write"}).get_text()
    days = bsObj.find("ul",{"class":"petitionsView_info_list"}).get_text()
    days = days.split("\n")
    category = days[1][4:]
    sdays = days[2][4:]
    edays = days[3][4:]
    person = days[4][3:8]
    count = "".join(re.findall("\d+", count))
    return [category, sdays, edays, person, progress, title, count, petition_overview]
```

```python
csvFile = open("2020_07_25_petition.csv", 'w', encoding='UTF-8')
writer = csv.writer(csvFile)
writer.writerow(["num","category","start-days","end-days","person","progress","title","count","petition_overview"])
for i in range(589941,589951):
    FindPetition(i) # 크롤링하는 함수 이름
csvFile.close()
```

```python
def FindPetition(num):
    url = "https://www1.president.go.kr/petitions/"
    url = url + str(num)
    try:
        html = urlopen(url)
    except urllib.error.HTTPError as e:
        print(e)
        print("=============error %d ============="%num)
    else:
        try:
            bsObj = BeautifulSoup(html.read(), "html.parser")
        except urllib.exceptions.SSLError as e:
            print(e)
            print("=============error %d ============="%num)
        else:
            try:
                progress = bsObj.find("div",{"class":"petitionsView_progress"}).get_text()
                title = bsObj.find("h3",{"class":"petitionsView_title"}).get_text()
                count = bsObj.find("h2",{"class":"petitionsView_count"}).get_text()
                petition_overview = bsObj.find("div",{"class":"View_write"}).get_text()
                days = bsObj.find("ul",{"class":"petitionsView_info_list"}).get_text()
                days = days.split("\n")
                category = days[1][4:]
                sdays = days[2][4:]
                edays = days[3][4:]
                person = days[4][3:8]
                count = "".join(re.findall("\d+", count))
            except:
                print("왜 안 되는지는 모르겠지(만 안 된다.")
            else:
                writer.writerow([category, sdays, edays, person, progress, title, count, petition_overview])
```
